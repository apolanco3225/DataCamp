```python
from urllib.request import urlopen, Request
# specify the url
url = 'https://www.eltiempo.com/'
# package the get request using the funtcion Request
request = Request(url)
# send the request and catch the response using urlopen
response = urlopen(request)
# apply the read method to the response whiche returns the
# html as a string 
html = response.read()
# close the response
response.close()
```

```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.crummy.com/software/BeautifulSoup/'
r = requests.get(url)
html_doc = r.text
soup = BeautifulSoup(html_doc)
```

```python
print(soup.title)
```

```python
print(soup.get_text())
```

```python
# extract the url of all the hyperlinks
for link in soup.find_all('a'):
    print(link.get('href'))
```
